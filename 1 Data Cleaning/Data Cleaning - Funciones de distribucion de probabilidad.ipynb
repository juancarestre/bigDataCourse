# Funciones de distribucion de probabilidad

## Distribucion Uniforme

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a=1
b=100
n=1000000
data=np.random.uniform(a,b,n)
```

```python
%matplotlib inline
plot=plt.hist(data)
```

## Distribucion normal

```python
data= np.random.randn(10000000)
```

```python
x=range(1,10000001)

plt.plot(x,data)
```

```python
plt.hist(data)
```

```python
plt.plot(x,sorted(data))
```

```python
mu= 5.5
sd = 2.5
data = mu + sd*np.random.randn(10000)
plt.hist(data)
```
